```python
def to_alpha(input):
    clean = ''
    for i in input:
        if i.isalpha() or i.isspace():
            clean = clean + i
    return clean
string = '%%%6clean c777at'
print(to_alpha(string))
```

```python
#function to only keep numeric characters in a string
def to_numeric(input):
    clean = ''
    for i in input:
        if i.isnumeric():
            clean = clean + i
    return clean
string = '%%%6clean c777at'
print(to_numeric(string))
```

```python
df['Description'] = df['Description'].apply(to_alpha)
df.head(2)
```
